# Usage of Function

### Define Function

````
def other_disease(x):
    if x == 1:
        x = '1st Class'
    if x == 2:
        x = '2nd Class'
    if x == 80:
        x = '3rd Class'
    return x
````

### Apply Function

````
df3['other_disease'] = df3['other_disease'].apply(lambda x: other_disease(x))
df3
````
